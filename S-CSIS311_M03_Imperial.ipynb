### Importing the libraries

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

### Importing the dataset

````python
dataset = pd.read_csv('D:\Luis\OneDrive - De La Salle University-Dasmari√±as\Documents\Reference Documents\BCS3 CSIS311 Reference Documents\Supplementary BCS3 CSIS311LA Reference Documents\Social_Network_Ads.csv')
````

````python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
#select and read age, salary X and y
````

### Splitting the dataset into the Training set and the Test set

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
#We get the same train and test sets across different executions
````

## Feature Scaling

````python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
#Age and Salary
````

````python
print(X_train)
````

### Training the K-NN model on the Training set

````python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
````

### Predicting a new result

````python
print(classifier.predict(sc.transform([[30,87000]])))
````

````python
print(classifier.predict(sc.transform([[43,36000]])))
````

````python
print(classifier.predict(sc.transform([[44,36000]])))
````

### Predicting the Test set Results

````python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1 ))
#not equal is incorrect 0,0 1,1 only
````

### Making the Confusion Matrix

## Recall and Precision

TP = 64 | FP = 4
--------|--------
FN = 3  | TN = 9

#### Accuracy

```
   TP + TN      = 64 + 9      73
 = ------------   ------    = ----- = 0.93
    TP+TN+FP+FN   64+9+4+3     80
```

#### Precision

```
    TP        64      64
 = -----  = -----   = --  = 0.9412 = 0.94 or 94%
   TP+FP    64 + 4     68
```

#### Recall

```
    TP       64
 = -----  = ---- = 0.9552 = 0.96
   TP+FN     67
```

#### F-1

```
    2 * precision * recall   2 * 0.9412 * 0.9552   1.8048
 = ----------------------- = ------------------- = ------ = 0.9481 = 0.95 or 95%
    precision + recall         0.9412 + 0.9552      1.9
```

````python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)

print("")
print("Accuracy: ", accuracy_score(y_test, y_pred))
````

````python
from sklearn.metrics import precision_score, recall_score, f1_score

print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("F-1: ", f1_score(y_test, y_pred))
````

````python
precision_score_manual = cm[0][0] / (cm[0][0] + cm[0][1])

print("Precision (manually computed): ", precision_score_manual)
````

````python
recall_score_manual = cm[0][0] / (cm[0][0] + cm[1][0])

print("Recall (manually computed): ", recall_score_manual)
````

````python
f1_score_manual = (2 * precision_score_manual * recall_score_manual) / (precision_score_manual + recall_score_manual)

print("F-1 (manually computed): ", f1_score_manual)
````

### Visualising the Training set results

````python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))

plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
    plt.title('K-NN (Training set)')
    plt.xlabel('Age')
    plt.ylabel('Estimated Salary')
    plt.legend()
plt.show()
````
